```python
import pandas as pd 
import numpy as np
```

```python
hou = pd.read_csv("KHOU.csv", sep=",")
nyc = pd.read_csv("KNYC.csv", sep=",")
sea = pd.read_csv("KSEA.csv", sep=",")
mdw = pd.read_csv("MDW.csv", sep=",")
phl = pd.read_csv("PHL.csv", sep=",")
phx = pd.read_csv("PHX.csv", sep=",")
```

```python
hou["date"]= pd.to_datetime(hou["date"])
hou['month']  = hou['date'].dt.strftime('%b')

nyc["date"]= pd.to_datetime(nyc["date"])
nyc['month']  = nyc['date'].dt.strftime('%b')

sea["date"]= pd.to_datetime(sea["date"])
sea['month']  = sea['date'].dt.strftime('%b')

mdw["date"]= pd.to_datetime(mdw["date"])
mdw['month']  = mdw['date'].dt.strftime('%b')

phl["date"]= pd.to_datetime(phl["date"])
phl['month']  = phl['date'].dt.strftime('%b')

phx["date"]= pd.to_datetime(phx["date"])
phx['month']  = phx['date'].dt.strftime('%b')
```

```python
phx.head()
```

```python
hou_high = hou.groupby('month', as_index=True)[['record_max_temp']].max().reset_index()
nyc_high = nyc.groupby('month', as_index=True)[['record_max_temp']].max().reset_index()
sea_high = sea.groupby('month', as_index=True)[['record_max_temp']].max().reset_index()
mdw_high = mdw.groupby('month', as_index=True)[['record_max_temp']].max().reset_index()
phl_high = phl.groupby('month', as_index=True)[['record_max_temp']].max().reset_index()
phx_high = phx.groupby('month', as_index=True)[['record_max_temp']].max().reset_index()
```

```python
hou_high.head(12)
```

```python
# Reorder the dates
months = {'Jan':1,'Feb':2,'Mar':3, 'Apr':4, 'May':5, 'Jun':6, 'Jul':7, 'Aug':8, 'Sep':9, 'Oct':10, 'Nov':11, 'Dec':12}

hou_high = hou_high.sort_values('month', key = lambda x : x.apply (lambda x : months[x])).reset_index(drop=True)
nyc_high = nyc_high.sort_values('month', key = lambda x : x.apply (lambda x : months[x])).reset_index(drop=True)
sea_high = sea_high.sort_values('month', key = lambda x : x.apply (lambda x : months[x])).reset_index(drop=True)
mdw_high = mdw_high.sort_values('month', key = lambda x : x.apply (lambda x : months[x])).reset_index(drop=True)
phl_high = phl_high.sort_values('month', key = lambda x : x.apply (lambda x : months[x])).reset_index(drop=True)
phx_high = phx_high.sort_values('month', key = lambda x : x.apply (lambda x : months[x])).reset_index(drop=True)
```

```python
# create new dataframe with all attributes and new month aggregation
df = pd.DataFrame()
df["month"] = hou_high["month"]
df["HOU"] = hou_high["record_max_temp"]
df["NYC"] = nyc_high["record_max_temp"]
df["SEA"] = sea_high["record_max_temp"]
df["MDW"] = mdw_high["record_max_temp"]
df["PHL"] = phl_high["record_max_temp"]
df["PHX"] = phx_high["record_max_temp"]
```

```python
df.to_csv("weather.csv", index=False)
```
